```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/COVID-19 AR Clinical Correlates July202020.csv")
print(df.shape)
df.head()
```

```python
df.drop(df.columns[[3, 4, 28]], axis=1, inplace=True)
```

```python
df.columns=df.iloc[0]
df=df[1:]
```

```python
df.head()
```

```python
df.set_index([df.columns[0]], inplace=True)
```

```python
df.dropna(inplace=True)
print(df.shape)
```

```python
height_list = list(df["LATEST HEIGHT"])
print(height_list)
```

```python
import re
def heightStr_to_cm(height_str):
    number = re.split('"|\'| ',height_str)
    number = [float(i) for i in number if i != ""]
    return number[0]*30.48 + number[1]*2.54

df["LATEST HEIGHT"] = [heightStr_to_cm(i) for i in height_list]
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler,LabelEncoder
```

```python
scale=MinMaxScaler()
df[['AGE','LATEST_BMI','LATEST WEIGHT','LATEST HEIGHT']]=scale.fit_transform(df[['AGE','LATEST_BMI','LATEST WEIGHT','LATEST HEIGHT']])
```

```python
df
```

```python
categ = ['SEX','TUBERCULOSIS', 'SYSTEMIC LUPUS ERYTHMATOSUS', 'RHEUMATOID ARTHRITIS',
       'EXTENSIVE BURNS', 'ASPLENIA', 'HYPOSPLENIA', 'MEASLES',
       'CYTOMEGALOVIRUS', 'CHICKEN POX', 'HERPES ZOSTER', 'MALNUTRITION',
       'CURRENT PREGNANT', 'CHRONIC KIDNEY DISEASE', 'DIABETES TYPE I',
       'DIABETES TYPE II', 'TRANSPLANT', 'HEMODIALYSIS Pre Diagnosis',
       'HEMODIALYSIS Post diagnosis', 'CANCER', 'COVID TEST POSITIVE',
       'ICU Admit', 'MORTALITY']

# Encode Categorical Columns
le = LabelEncoder()
df[categ] = df[categ].apply(le.fit_transform)
```

```python
df
```

```python
df["MORTALITY"].value_counts()
```

```python
num_col=['AGE','LATEST_BMI','LATEST WEIGHT','LATEST HEIGHT','MORTALITY']
num_data=df[num_col]
```

```python
num_data
```

```python
from sklearn.model_selection import train_test_split
X=num_data[['AGE','LATEST_BMI','LATEST WEIGHT','LATEST HEIGHT']]  # Features
y=num_data['MORTALITY']  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier

RF_clf =RandomForestClassifier(n_estimators=100)

RF_clf.fit(X_train,y_train)

y_pred=RF_clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
DT_clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = DT_clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(X_train,y_train)

#Predict Output
y_pred= model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
